# Feature Scaling (Standardiation) :     


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/anjali801/AI-and-Machine-Learning-/main/drama1.csv"
df = pd.read_csv(url)
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df["Content Rating"] = df["Content Rating"].fillna(df["Content Rating"].mode()[0])
```

```python
sns.distplot(df["Year of release"])
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(df[["Year of release"]])
```

```python
ss.transform(df[["Year of release"]])
```

```python
df["Year of release _ss"] = pd.DataFrame(ss.transform(df[["Year of release"]]))
```

```python
df
```

```python
df.describe()
```

```python
plt.subplot(1,2,1)
plt.title("After Scaling")
sns.distplot(df["Year of release _ss"])

plt.subplot(1,2,2)
plt.title("Before Scaling")
sns.distplot(df["Year of release"])
```
