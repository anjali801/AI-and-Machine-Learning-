# Ordinal Encoding :                                        
do using :   ------>sklearn ------->map function               

```python
import pandas as pd
df = pd.DataFrame({"Size" : ["s","m","l","xl","s","s","l","xl"]})
df
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=ord_data)
oe_e = oe.fit(df[['Size']])
print(oe_e)
```

```python
oe_e1 = oe_e.transform(df[['Size']])
print(oe_e1)
```

```python
df["size_oe"] = oe_e1             #  show ki kisko konsi value mili hai
df
```

```python
# through map functiom :    ------
ord_data1 = {"s" : 0 ,"m" : 1 ,"l":2,"xl":3}
```

```python
df["Size"].map(ord_data1)
```

```python
df["Size_en_map"]=df["Size"].map(ord_data1)               # for save the data
```

#outlier removal using IQR:   

IQR ----> Inter qutlier range

```python
dataset.shape
```

```python
q1 = dataset["physics"].quantile(0.25)
q3 = dataset["physics"].quantile(0.75)
print(q1)
print(q3)
```

```python
IQR = q3 - q1
```

```python
min_range = q1 - (1.5 * IQR)
max_range = q3 + (1.5 * IQR)
min_range , max_range
```

```python
# oridinal data se outlier hatana hai :
new_dataset = dataset[dataset["physics"]<= max_range]
```
