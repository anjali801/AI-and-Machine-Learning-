#Encoding :
categariol data convert into numerical data .


---



--> one hot encoding :------------------------------ use : when short number of the data is present (ex - male & female , Yes & No)

```python
import pandas as pd
data = {
    "name" : ["A","B","C","D","E","F","G"],
    "gender" : ["male","female","male","female","male","female","male"]
}
df = pd.DataFrame(data)
print(df)
```

```python
df.isnull().sum()
```

# : encoding by get dummies

```python
pd.get_dummies(df['gender'])            # check by info then shows  boolean data
```

# : encoding by OneHotEncoder class

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe_e = ohe.fit_transform(df[['gender']])         # give sparse matrix
print(ohe_e)
```

```python
ohe_e1 = ohe_e.toarray()
print(ohe_e1)
```

```python
pd.DataFrame(ohe_e1,columns = ohe.categories_)
```
